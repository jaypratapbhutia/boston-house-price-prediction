```python
import pandas as pd
from sklearn import datasets   # to import the dataset of boston present in sklearn 
from sklearn.model_selection import train_test_split # to split the train and test dataset
from sklearn.linear_model import LinearRegression # to find the linear regression
from sklearn.metrics import mean_squared_error #to find the error in prediction
```

```python
boston = datasets.load_boston() # importing the boston dataset from the scikitlearn
```

```python
print(boston.DESCR) #describes the dataset
```

value of X=13,  y = medv ,
number of observation=n=506 ,

X=feature columns or independent variable ,
y=target or dependent variable


```python
X=pd.DataFrame(boston.data,columns=boston.feature_names)
y=pd.DataFrame(boston.target,columns=['MEDV'])

X #prints the FEATURE values of X 
```

```python
y # PRINTS THE TARGET VALUE 
```

```python
X.isnull().sum() # used for checking for null values in X dataset
```

```python
y.isnull().sum() # used to check for null values in y dataset
```

```python
# used to split the data set into train,test pair
# train =70%,test=30%

train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
# FIND THE REGRESSION MODEL OFTHE GIVEN TRAIN DATA
reg_model=LinearRegression().fit(train_x,train_y)
```

```python
# predicting the target value using the test dataset
pred_y= reg_model.predict(test_x)
```

```python
pred_y
```

```python
# finding the error as compared to the y dataset
print(mean_squared_error(test_y,pred_y))
#or
import numpy as np
np.mean((pred_y-test_y)**2)
```

```python
# finding the regression model for the original dataset
print(reg_model.score(train_x,train_y))
```

### R-squared >=0.85 -BEST FIT
### R-squared>=0.7 -GOOD FIT
### R-squared>=0.5 -poor fit
 as the value of r here is .74 so this is good fit model
